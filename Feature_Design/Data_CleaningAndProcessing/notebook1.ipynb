### **Handling Outliers**

```python
raw_data.info()
```

```python
raw_data.describe()
```

Information deduced from these tables have shown that there values that can be considered as outliers. let me identify the outliers and remove them using interquartile range `(iqr)`

```python
from scipy.stats import iqr
import numpy as np

# Make a copy to keep original safe
df = raw_data.copy()


mask = np.ones(len(df), dtype=bool)
col_list = raw_data.columns

for col in col_list:
    data = df[col].values
    col_iqr = iqr(data)
    q1 = np.quantile(data, 0.25)
    q3 = np.quantile(data, 0.75)
    lower_threshold = q1 - 1.5 * col_iqr
    upper_threshold = q3 + 1.5 * col_iqr
    
    # Update mask
    mask &= (data >= lower_threshold) & (data <= upper_threshold)

# Apply mask once
clean_raw_data = df[mask]

print("Original shape:", raw_data.shape)
print("After removing outliers:", clean_raw_data.shape)
```

This shows that 94 enteries have been identified as outliers and removed.

### **Distribution of the Data**

```python
import seaborn as sns

columns = clean_raw_data.columns

# Number of features
n = len(columns)
rows = (n // 3) + 1
cols =3

plt.figure(figsize=(15, 4*rows))

for i, col in enumerate(columns, 1):
    if col == 'water_cement_ratio':
        break
    
    plt.subplot(rows, cols, i)
    sns.histplot(clean_raw_data[col], kde=True, bins=30, color='steelblue')
    plt.title(f'{col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    

plt.tight_layout()
plt.show()
```

This plot shows that the distribution of the data can be approxiated to normal which is suitable for training machine learning models.

### **Engineering-Based Feature Design**

In this section, four engineering features were introduced to the dataset namely:
- water-cement ratio
- Binder content
- water-binder ratio
- Aggregate ratio

$$
water-cement-ratio (w/c)= water/cement 
$$
$$
binder = cement + slag + flyash
$$
$$
water-binder-ratio (w/b) = water/binder
$$
$$
aggregate-ratio = fine aggregate/coarse aggregate
$$

```python
#copy the raw data and introduce the engineering features
eng_data = clean_raw_data.copy()
eng_data['w_c'] = eng_data['water'] / eng_data['cement']
eng_data['binder'] = eng_data['cement'] + eng_data['blast_furnace_slag'] + eng_data['fly_ash']
eng_data['w_b'] = eng_data['water'] / eng_data['binder']
eng_data['agg_ratio'] = eng_data['fine_aggregate'] / eng_data['coarse_aggregate']

eng_fea = eng_data[["w_c", "binder", "w_b", "agg_ratio"]]
eng_fea.head()
```

```python
import seaborn as sns

columns = eng_fea.columns

# Number of features
n = len(columns)
rows = (n // 4) + 1
cols = 2

plt.figure(figsize=(10, 2*rows))

for i, col in enumerate(columns, 1):
    #if col == 'water_cement_ratio':
        #break
    
    plt.subplot(rows, cols, i)
    sns.histplot(eng_fea[col], kde=True, bins=30, color='steelblue')
    #plt.title(f'{col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    

plt.tight_layout()
plt.show()
```
